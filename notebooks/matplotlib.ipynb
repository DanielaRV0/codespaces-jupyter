```python
import matplotlib.pyplot as plt

# Datos
tiempo = [0, 10, 20, 30, 40, 50]
posicion = [30, 52, 38, 0, -37, -53]

# Crear la gráfica
plt.plot(tiempo, posicion, marker='o', linestyle='-', color='pink')

# Etiquetas de los ejes
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')

# Título
plt.title('Gráfico de Posición vs Tiempo')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
